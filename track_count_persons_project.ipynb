```python
import matplotlib.pyplot as plt
from sklearn.metrics import precision_recall_curve, average_precision_score, recall_score, accuracy_score

import cv2
import os
from xml.etree import ElementTree

import tensorflow as tf
from sklearn.metrics import confusion_matrix
from tensorflow.keras import datasets, layers, models
keras = tf.keras
```

```python
class_names = ['person','person-like']
class_names_label = {class_name:i for i, class_name in enumerate(class_names)}

n_classes = 2
size = (120,120)
```

```python
def load_data():
    datasets = ['/content/drive/MyDrive/SEM_6/PROJECTs/capstone_files/Pedestrian_Detection/Train/Train',
               '/content/drive/MyDrive/SEM_6/PROJECTs/capstone_files/Pedestrian_Detection/Test/Test',
               '/content/drive/MyDrive/SEM_6/PROJECTs/capstone_files/Pedestrian_Detection/Val/Val']
    output = []

    for dataset in datasets:
        imags = []
        labels = []
        directoryA = dataset + "/Annotations"
        directoryIMG = dataset + "/JPEGImages/"
        file = os.listdir(directoryA)
        img = os.listdir(directoryIMG)
        file.sort()
        img.sort()

        # Check for unequal list lengths and handle gracefully
        if len(file) != len(img):
            raise ValueError("Unequal number of XML and image files. Consider handling the difference appropriately (e.g., skipping extra XML files or logging errors).")

        i = 0
        for xml in file:

            xmlf = os.path.join(directoryA, xml)
            dom = ElementTree.parse(xmlf)
            vb = dom.findall('object')
            label = vb[0].find('name').text
            labels.append(class_names_label[label])

            # Check for empty img list and handle accordingly
            if not img:
                raise ValueError("Empty image list. This might indicate missing image files or an error in directory structure.")

            # Access img[i] only if it exists
            if i < len(img):
                img_path = directoryIMG + img[i]
                curr_img = cv2.imread(img_path)
                curr_img = cv2.resize(curr_img, size)
                imags.append(curr_img)

            i += 1

        imags = np.array(imags, dtype='float32')
        imags = imags / 255

        labels = np.array(labels, dtype='int32')

        output.append((imags, labels))
    return output
```

```python
(train_images, train_labels),(test_images, test_labels),(val_images, val_labels) = load_data()
```

```python
import matplotlib.pyplot as plt
from sklearn.metrics import precision_recall_curve, average_precision_score, recall_score, accuracy_score

# Assuming you have already loaded your train and test data using the `load_data` function
# and made predictions on both using the YOLO model

# Define functions for calculating and plotting metrics

def plot_precision_recall(y_true, y_score, label):
  """
  Plots the precision-recall curve for the given data.

  Args:
      y_true: True labels for the data.
      y_score: Predicted probabilities for the data.
      label: Label for the plot (e.g., "Train", "Test").
  """
  precision, recall, _ = precision_recall_curve(y_true, y_score)
  average_precision = average_precision_score(y_true, y_score)

  plt.step(recall, precision, where='post', label=f"{label} (average precision = {average_precision:.2f})")
  plt.xlabel('Recall')
  plt.ylabel('Precision')
  plt.title('Precision-Recall Curve')
  plt.legend()

def plot_other_metrics(y_true, y_pred, label):
  """
  Calculates and prints other evaluation metrics for the given data.

  Args:
      y_true: True labels for the data.
      y_pred: Predicted labels for the data.
      label: Label for the data (e.g., "Train", "Test").
  """
  accuracy = accuracy_score(y_true, y_pred)
  recall = recall_score(y_true, y_pred)

  print(f"{label} Accuracy: {accuracy:.4f}")
  print(f"{label} Recall: {recall:.4f}")

# Calculate and handle potential tensor dimensions and empty cases
if isinstance(train_predictions, list):
  train_predictions = np.array(train_predictions)  # Convert list to NumPy array if necessary

if train_predictions.shape[1] > 0:  # Check if there are any columns in train_predictions
  if len(train_predictions.shape) == 2:  # If 2D (batch size x classes)
    train_scores = np.squeeze(train_predictions[:, 1])
  else:                                 # If more dimensions (e.g., batch size x height x width x classes)
    train_scores = train_predictions[..., 1]  # Extract the second channel from the last dimension
else:
  print("Warning: train_predictions has no columns, skipping evaluation for train data.")
  train_scores = None  # Set train_scores to None to avoid NameError

if isinstance(test_predictions, list):
  test_predictions = np.array(test_predictions)  # Convert list to NumPy array if necessary

if test_predictions.shape[1] > 0:  # Check if there are any columns in test_predictions
  if len(test_predictions.shape) == 2:  # If 2D (batch size x classes)
    test_scores = np.squeeze(test_predictions[:, 1])
  else:                                 # If more dimensions (e.g., batch size x height x width x classes)
    test_scores = test_predictions[..., 1]  # Extract the second channel from the last dimension
else:
  print("Warning: test_predictions has no columns, skipping evaluation for test data.")
  test_scores = None  # Set test_scores to None to avoid NameError

# Skip evaluation if both train_scores and test_scores are None
if train_scores is None and test_scores is None:
  print("Skipping evaluation as both train_predictions and test_predictions have no columns.")
else:
  # Calculate and plot precision-recall curves for train and test data (if applicable)
  if train_scores is not None:
    plot_precision_recall(train_labels, train_scores, "Train")
  if test_scores is not None:
    plot_precision_recall(test_labels, test_scores, "Test")
  plt.show()

  # Calculate and print other metrics for train and test data (if applicable)
  if train_scores is not None:
    plot_other_metrics(train_labels, np.argmax(train_predictions, axis=1), "Train")
  if test_scores is not None:
    plot_other_metrics(test_labels, np.argmax(test_predictions, axis=1), "Test")
```
